# Подготовка датасета

```python
import datetime

def get_timestamp(time: str, sep="-"):
    time = str(time)
    stime = list(map(int, time.split(sep))) # 2023-02-20
    return int(datetime.datetime(stime[0], stime[1], stime[2]).timestamp())

```

Выделим заказы, про которые мы знаем закончены ли они в срок или нет, т.е. 100 !=  ПроцентЗавершенияЗадачи или ДатаокончанияБП0 is Nan

```python
fineshed = data[data["ПроцентЗавершенияЗадачи"] == 100]
fineshed
```

```python
ended = fineshed[fineshed["ДатаокончанияБП0"].notna()]
ended = ended[ended["ДатаначалаБП0"].notna()]

ended
```

Возьмём только последние события с объектом

```python
ended["ДатаНачалаЗадачи"] = ended["ДатаНачалаЗадачи"].apply(lambda x: get_timestamp(x))
ended["ДатаОкончанияЗадачи"] = ended["ДатаОкончанияЗадачи"].apply(lambda x: get_timestamp(x))
ended["ДатаначалаБП0"] = ended["ДатаначалаБП0"].apply(lambda x: get_timestamp(x))
ended["ДатаокончанияБП0"] = ended["ДатаокончанияБП0"].apply(lambda x: get_timestamp(x))
```

```python
idx = ended.groupby(['obj_key', 'Кодзадачи'])['ДатаокончанияБП0'].transform(max) == ended['ДатаокончанияБП0']
clean_data = ended[idx]
```

Сделаем лейбл max([ДатаОкончанияЗадачи - ДатаокончанияБП0], 0)



```python
clean_data['label'] = clean_data['ДатаОкончанияЗадачи'] - clean_data['ДатаокончанияБП0']
clean_data['label'] = clean_data['label'].apply(lambda x: max(x, 0) // 86400)
clean_data
```

```python
clean_data['label'].hist(bins=100)
```

Будем считать, что откланение от графика на более 180 дней (полгода) - это максимум, далее предсказывать будет очень сложно  

```python
clean_data['label'] = clean_data['label'].apply(lambda x: min(180, x))
clean_data['label'].hist(bins=100)
```

Добавим аттрибуты

```python
clean_data = pd.merge(clean_data, data_att, how="left", on=["obj_key", "date_report"])
clean_data
```

Посплитим на трейн и тест по объектам

```python
from sklearn.model_selection import train_test_split

objs = list(set(clean_data["obj_key"].values))
objs_train, objs_test = train_test_split(objs, test_size=0.33)
print(len(objs_train), len(objs_test))
```

```python
train_data = clean_data.loc[clean_data['obj_key'].isin(objs_train)]
train_data
```

```python
test_data = clean_data.loc[clean_data['obj_key'].isin(objs_test)]
test_data
```

оставим нужные поля

```python
import numpy as np
import math
from sklearn.preprocessing import RobustScaler

used_feat = ["obj_key", "Кодзадачи",
             "Статуспоэкспертизе", "Экспертиза",
             "состояние площадки", "Площадь", "Генпроектировщик",
             "Генподрядчик",	"Кол-во рабочих"]

train_data["Статуспоэкспертизе"] = train_data["Статуспоэкспертизе"].apply(lambda x: 0 if math.isnan(x) else int(x))
test_data["Статуспоэкспертизе"] = test_data["Статуспоэкспертизе"].apply(lambda x: 0 if math.isnan(x) else int(x))

train_data_nl, test_data_nl = train_data[used_feat], test_data[used_feat]

# rs = RobustScaler().fit([train_label])

# train_label, test_label = rs.transform([train_data["label"]])[0], rs.transform([test_data["label"]])[0]
train_label, test_label = train_data["label"], test_data["label"]
train_label
```

```python
from catboost import Pool, CatBoostRegressor

cat_features = [
    "obj_key", "Кодзадачи",
    "состояние площадки", "Статуспоэкспертизе",
    "Экспертиза",
  ]

for cf in cat_features:
  train_data_nl[cf], test_data_nl[cf] = train_data_nl[cf].fillna("Nan"), test_data_nl[cf].fillna("Nan")

train_dataset = Pool(data=train_data_nl,
                     label=train_label,
                     cat_features=cat_features)

eval_dataset = Pool(data=test_data_nl,
                    label=test_label,
                    cat_features=cat_features)

model = CatBoostRegressor(iterations=100,
                          learning_rate=0.01,
                          loss_function="RMSE")


model.fit(train_dataset,
          eval_set=eval_dataset)
```
